```python
import glob
import matplotlib.pyplot as plt
import numpy as np

def getUserItemMatrix(line):
    line = line.strip()
    rowList = [int(x) for x in line.rstrip('\n').split(" ")]
    return(rowList[0])

userItem=[]
for name in glob.glob('rec-list.dat'):
    for line in open(name,'r+'):
        temp = line.strip()
        stArray = temp.split(":")
        userItem.append(int(stArray[0]))

totalDocsPerUser=[]
for line in open('../ordered-cf-test-1-users.dat', 'r+'):
    totalDocsPerUser.append(getUserItemMatrix(line))


precision = [x/10 for x in userItem]
recall=[]
loop=0
for x in userItem:
    if(totalDocsPerUser[loop]!=0):
        recall.append(x/(totalDocsPerUser[loop]))
    else:
            recall.append(0)
            
    loop = loop+1

```

```python
# users with more than 5 papers in the test library
precision1 = precision[1:1834]

# users with 5 or less than 5 in the test library
precision2 = precision[1835:5551]
plt.figure(1, figsize=(20, 6))
plt.plot(precision1)
plt.ylabel('precision')
plt.xlabel('users in decreasing order of library size ->' + '\n' +'All users have more than 5 docuemnts in their test library')
plt.show()

plt.figure(2, figsize=(20, 6))
p = np.sort(precision1)
plt.xticks(np.arange(0, len(p), step=100))
plt.ylabel('precision')
plt.xlabel('users in increasing order of precision ->' + '\n' +'All users have more than 5 docuemnts in their test library')
plt.grid(True)
plt.plot(p)
plt.show()

plt.figure(3, figsize=(20, 6))
plt.plot(precision2)
plt.ylabel('precision')
plt.xlabel('users in decreasing order of library size ->' + '\n' +'All users have less than 5 docuemnts in their test library')
plt.show()

plt.figure(4, figsize=(20, 6))
p = np.sort(precision2)
plt.plot(p)
plt.ylabel('precision')
plt.xlabel('users in increasing order of precision ->' + '\n' +'All users have less than 5 docuemnts in their test library')
plt.xticks(np.arange(0, len(p), step=100))
plt.grid(True)
plt.show()
```

```python

# users with more than 5 papers in the test library
recall1 = recall[1:1834]

# users with 5 or less than 5 in the test library
recall2 = recall[1835:5551]
plt.figure(1, figsize=(20, 6))
plt.plot(recall1)
plt.ylabel('recall')
plt.xlabel('users in decreasing order of library size ->' + '\n' +'All users have more than 5 docuemnts in their test library')
plt.show()

plt.figure(2, figsize=(20, 6))
p = np.sort(recall1)
plt.xticks(np.arange(0, len(p), step=100))
plt.ylabel('recall')
plt.xlabel('users in increasing order of recall ->' + '\n' +'All users have more than 5 docuemnts in their test library')
plt.grid(True)
plt.plot(p)
plt.show()

plt.figure(3, figsize=(20, 6))
plt.plot(recall2)
plt.ylabel('recall')
plt.xlabel('users in decreasing order of library size ->' + '\n' +'All users have less than 5 docuemnts in their test library')
plt.show()

plt.figure(4, figsize=(20, 6))
p = np.sort(recall2)
plt.plot(p)
plt.ylabel('recall')
plt.xlabel('users in increasing order of recall ->' + '\n' +'All users have less than 5 docuemnts in their test library')
plt.xticks(np.arange(0, len(p), step=100))
plt.grid(True)
plt.show()
```
