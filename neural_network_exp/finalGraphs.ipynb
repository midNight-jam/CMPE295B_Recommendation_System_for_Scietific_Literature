```python
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
import numpy as np

rd1 = [0.007205908845, 0.006809583859, 0.00698973158]
rd2 = [0.04545727497, 0.02950819672, 0.01540263016]
rd3 = [0.001741427971, 0.001693388579, 0.00127904882]
rd4 = [0.2137752957, 0.2046838407, 0.1757701315]

plt.figure(1, figsize=(8, 4))
plt.plot(rd1, 'b', marker='v', linestyle='dashed', linewidth=2, markersize=6)
plt.plot(rd2, 'g', marker='v', linestyle='dashed', linewidth=2, markersize=6)
plt.plot(rd3, 'r', marker='v', linestyle='dashed', linewidth=2, markersize=6)
plt.plot(rd4, 'm', marker='o', linestyle='dashed', linewidth=2, markersize=6)

c1 = mpatches.Patch(color='b', label='Content ML')
c2 = mpatches.Patch(color='g', label='AutoEncoSVD')
c3 = mpatches.Patch(color='r', label='CollabML')
c4 = mpatches.Patch(color='m', label='CDL')
plt.legend(handles=[c1, c2, c3, c4], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('M', weight = 8) 
plt.ylabel('Precision')
plt.title('Precision @ M')
plt.xticks([0, 1, 2], ['M = 3', 'M = 5', 'M = 10'])
plt.yticks(np.arange(0, 0.30, step=0.025))
plt.grid(True)
plt.show()
```
