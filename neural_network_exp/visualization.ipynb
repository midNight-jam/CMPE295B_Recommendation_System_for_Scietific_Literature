```python
import numpy as np
import matplotlib.pyplot as plt
import data_visualization as data

mat = data.read_and_create_user_LibCount()

plt.figure(1, figsize=(15, 6))
plt.xlabel('Users')
plt.ylabel('Papers in library')
plt.title('Number of papers in users library')
plt.plot(mat)

plt.figure(2, figsize=(15, 6))
plt.xticks(rotation=55)
binwidth = 5
bins=np.arange(min(mat), max(mat) + binwidth, binwidth)
plt.hist(mat,bins, color='c', edgecolor='black', linewidth=0.2)
plt.xlabel('Papers in library')
plt.ylabel('Users')
plt.grid(True)
plt.yticks(np.arange(0, 1600, step=100))
plt.title('UserLibrary Size Distribution Histogram'.format(binwidth))
plt.show()
```
